```python
df_new = pd.read_csv(url, parse_dates=['ObservationDate', 'Last Update'])
df_new.dtypes
```

```python
def corrige_colunas(col_name):
    return re.sub(r"[/| ]", "", col_name).lower()
```

```python
df_new.columns = [corrige_colunas(col) for col in df_new.columns]
df_new
```

```python
brasil_new = df_new.loc[
    (df_new.countryregion == 'Brazil') & 
    (df_new.confirmed > 0)
]
brasil_new
```

```python
brasil_new = brasil_new.groupby('observationdate').sum()
brasil_new
```

```python
brasil_new = brasil_new.reset_index()
brasil_new
```

```python
filtro = brasil_new['observationdate'] < '2020-12-01'
brasil_new = brasil_new[filtro]
brasil_new.dtypes
```

```python
exportar = brasil_new.to_csv("covid_brasil_2020.csv", index=False)
exportar
```

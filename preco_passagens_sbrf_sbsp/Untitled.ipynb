```python
import pandas as pd
```

```python
df = pd.read_csv('flights_sbrf_to_sbsp.csv')
```

```python
tarifas = df['TARIFA']
for i in range(len(tarifas)):
    df['TARIFA'][i] = df['TARIFA'][i].replace(',','.')
df
```

```python
df.TARIFA = df.TARIFA.astype(float)
print(df.dtypes)
```

```python
df.rename(columns={'ANO':'ano','MES':'mes','EMPRESA':'empresa','ORIGEM':'origem','DESTINO':'destino','TARIFA':'tarifa','ASSENTOS':'assentos'}, inplace=True)
df
```

```python
df['data'] = ''
for i in range(len(df)):
    if df['mes'][i] < 10:
        df['data'][i] = f"{df['ano'][i]}-0{df['mes'][i]}-01"
    else:
        df['data'][i] = f"{df['ano'][i]}-{df['mes'][i]}-01"
```

```python
df['data'] = pd.to_datetime(df['data'])
print(df.dtypes)
```

```python
df
```

```python
df_mes = df.groupby('data').sum()
```

```python
df_mes.drop(['ano','mes'], axis=1, inplace=True)
```

```python
df_mes['tarifaporassento'] = ''
for c in range(len(df_mes)):
    df_mes['tarifaporassento'][c] = df_mes['tarifa'][c]/df_mes['assentos'][c]
df_mes['tarifaporassento'] = pd.to_numeric(df_mes['tarifaporassento'])
round(df_mes["tarifaporassento"], 2)
df_mes
```

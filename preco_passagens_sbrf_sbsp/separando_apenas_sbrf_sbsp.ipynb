```python
import pandas as pd
```

```python
dicionario = {'2017': ['08','09','10','11','12'], 
              '2018': ['01','02','03','04','05','06','07','08','09','10','11','12'],
              '2019': ['01','02','03','04','05','06','07','08','09','10','11','12'],
              '2020': ['01','02','03','04','05','06','07','08','09','10','11','12'],
              '2021': ['01','02','03','04','05','06','07','08','09','10','11','12'],
              '2022': ['01','02','03','04','05','06','07']
}
lista_de_arquivos = []
sbrf_sbsp = pd.DataFrame(columns=['ANO','MES','EMPRESA','ORIGEM','DESTINO','TARIFA','ASSENTOS'])
for k, l in dicionario.items():
    for v in range(len(l)):
        lista_de_arquivos.append(f'{k}{l[v]}.CSV')
```

```python
def lista_voos(nome):
    lista_acumulada = []
    lista_erro = []
    try:
        temporario = pd.read_csv(nome, sep=';')
        for linha in range(len(temporario)):
            temp_list = (temporario.iloc[linha]).to_list()
            if temp_list[3] == 'SBRF' and temp_list[4] == 'SBSP':
                lista_acumulada.append(temp_list)
    except UnicodeDecodeError:
        lista_erro.append(nome)
        print(lista_erro)
    else:
        return lista_acumulada
```

```python
for cont in range(len(lista_de_arquivos)):
    lista_com_voos = lista_voos(lista_de_arquivos[cont])
    for i in range(len(lista_com_voos)):
        sbrf_sbsp.loc[len(sbrf_sbsp)] = lista_com_voos[i]
```
